- # Gráfico de pizza

```python
import matplotlib.pyplot as plt

# Dados do gráfico de pizza
labels = ['A', 'B', 'C', 'D']  # Nomes das categorias
sizes = [15, 30, 45, 10]  # Tamanhos ou valores das categorias
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Cores para cada fatia
explode = (0.1, 0, 0, 0)  # "Explodir" a primeira fatia (destaque)

# Criar o gráfico de pizza
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Título do gráfico
plt.title('Gráfico de Pizza Exemplo')

# Exibir o gráfico
plt.axis('equal')  # Para garantir que o gráfico de pizza seja desenhado como um círculo
plt.show()
```

- # Gráfico de dispersão

```python
import matplotlib.pyplot as plt

# Dados de exemplo
x = [1, 2, 3, 4, 5]  # Eixo X (variável independente)
y = [2, 3, 5, 7, 11]  # Eixo Y (variável dependente)

# Criar o gráfico de dispersão
plt.scatter(x, y, color='blue', marker='o')  # Define a cor e o marcador das bolinhas

# Adicionar títulos aos eixos e ao gráfico
plt.title('Gráfico de Dispersão: Exemplo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Exibir o gráfico
plt.show()
```

- # Grafico bloxplot

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
dados = np.random.normal(0, 1, 100)  

plt.boxplot(dados)

plt.title('Gráfico de Boxplot com Matplotlib')
plt.ylabel('Valor')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gerar alguns dados de exemplo (dados aleatórios)
np.random.seed(10)
dados = np.random.normal(0, 1, 100)  # 100 números com distribuição normal

sns.boxplot(data=dados)

plt.title('Gráfico de Boxplot com Seaborn')
plt.ylabel('Valor')

plt.show()
```
