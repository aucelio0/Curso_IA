### O que é IA?

Inteligência Artificial (IA) é um ramo da computação que cria sistemas capazes de simular comportamentos humanos inteligentes. Isso inclui habilidades como:

Aprender com dados, até se tornar com conhecimento 

Raciocinar e resolver problemas

Reconhecer padrões (como rostos ou vozes)

Compreender linguagem natural

Tomar decisões automatizadas

### Cargas de trabalho comuns com IA?

Análise e previsões de dados

Automação

Visão computacional

### Princípios de IA responsável

Justiça

Transparência

Privacidade e segurança

Inclusão

Confiabilidade

### O que é aprendizado da máquina

Aprendizado de máquina é uma área da Inteligência Artificial onde as máquinas aprendem sozinhas a fazer tarefas a partir de dados, sem serem programadas linha por linha pra cada coisa.

### Tipos de aprendizado de máquina

Aprendizado supervisionado
Você dá os dados com as respostas certas.
Ex: Fotos + rótulo (“gato” ou “cachorro”).

Aprendizado não supervisionado
A máquina descobre padrões sozinha, sem respostas dadas.
Ex: Agrupar clientes parecidos com base em hábitos.

Aprendizado por reforço
A máquina aprende com tentativa e erro, como um jogo.
Ex: Treinar um robô ou IA pra jogar xadrez.

### Treinamento e validação de modelo

- ##### Treinamento
    Fornecer um conjunto de dados com as respostas certas.
    O modelo ajusta seus “pesos” internos para tentar prever essas respostas.
    Quanto mais exemplos, mais aprende os padrões

- ##### Validação
    É quando a IA é testada enquanto ainda está aprendendo — mas com dados diferentes dos que viu no treino.
    erve pra verificar se o modelo está realmente aprendendo, e não só decorando os dados.
    Ajuda a ajustar os parâmetros do modelo (como a taxa de aprendizado ou a estrutura da rede neural)

### O que é aprendizado profundo? 

Imagine uma rede neural como um "cérebro artificial", com camadas de neurônios conectados entre si. O aprendizado profundo usa redes com várias camadas, e cada camada aprende algo diferente dos dados. As camadas mais profundas aprendem padrões mais complexos.

Exemplo: Reconhecimento de Imagens
Primeira camada: Detecta bordas e cores simples.

Segunda camada: Junta as bordas e forma formas básicas (círculos, quadrados).

Camadas mais profundas: Junta essas formas e identifica características mais avançadas (olhos, nariz, etc.).


-   ##### O que torna o aprendizado profundo diferente?
    Enquanto o aprendizado de máquina tradicional usa modelos simples (como regressão ou máquinas de vetor de suporte), o aprendizado profundo trabalha com redes neurais profundas, que são compostas por muitas camadas de nós (ou "neurônios"), inspirados no funcionamento do cérebro humano. Essas redes têm muitas camadas (daí o "profundo"), o que permite que o sistema aprenda representações mais complexas dos dados.
