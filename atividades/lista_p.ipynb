```python
# 6. Desenvolva um programa que pergunte ao usuário uma operação matemática (+, -, *, /) e dois números, e realize a operação escolhida.
 
while True:   
    # operações
    print(f'{'='*30} Operações {'='*30}\n')
    print('1 - SOMA')
    print('2 - SUBTRAÇÃO')
    print('3 - MULTIPLICAÇÃO')
    print('4 - DIVISÃO')

    # seleção da operação
    opcao = input('\nEscolha qual operação: ')

    x = str(input('Informe um número: ')).replace(',', '.')
    y = str(input('Informe outro número: ')).replace(',', '.')
    
    # conversão
    x = float(x)
    y = float(y)

    match int(opcao):
        case 1:
            try:
                soma = x + y
                print(f'Resultado:{soma}')
            except:
                print('Não foi possível realizar cálculo.')
        case 2:
            try:
                subtracao = x - y
                print(f'Resultado:{subtracao}') 
            except:
                print('Não foi possível realizar cálculo.')  
        case 3: 
            try: 
                multiplicacao = x * y
                print(f'Resultado:{multiplicacao}') 
            except:
                print('Não foi possível realizar cálculo.')  
        case 4:
            try:
                divisao = x / y
                print(f'Resultado:{divisao}') 
            except:
                print('Não foi possível realizar cálculo.')  
        
 # continuar com algum cálculo
     
    continuar = input('Deseja continuar (s/n)').lower()

    if continuar == 's':
      continue
    else:
       break
       
```

```python
# 44 Crie um programa que peça ao usuário 10 números e exiba apenas os números pares.

numeros = []

for i in range(10):
    while True:
        try:
            numero = int(input(f"Digite o {i + 1}º número: "))
            numeros.append(numero)
            break
        except ValueError:
            print("Digite um número inteiro.")

pares = [n for n in numeros if n % 2 == 0]

print("\nNúmeros pares digitados:")
if pares:
    for par in pares:
        print(par)
else:
    print("Nenhum número par foi digitado.")
```

```python
# 56 Escreva um programa que pergunte ao usuário quantas vezes ele quer que uma mensagem seja exibida, e utilize um laço while para exibir a mensagem a quantidade de vezes desejada.

while True:
    try:
        vezes = int(input("Quantas vezes você quer que a mensagem seja exibida? "))
        if vezes > 0:
            break 
        else:
            print("Insira um número maior que 0.")
    except ValueError:
        print("Digite um número inteiro válido.")

contador = 0
while contador < vezes:
    print("Esta é a sua mensagem!")
    contador += 1
```

```python
# 57 Crie um programa que peça ao usuário para adivinhar um número secreto entre 1 e 10. Continue pedindo até que ele adivinhe o número corretamente.
import random

numero = random.randint(1, 10)

while True:
    try:
        palpite = int(input("Adivinhe o número secreto entre 1 e 10: "))
        
        if palpite == numero:
            print("Você acertou o número secreto!")
            break 
        elif palpite < numero:
            print("Tente um número maior.")
        else:
            print("Tente um número menor.")
    except ValueError:
        print("Digite um número válido.")
```

```python
# 59 Escreva um programa que solicite ao usuário para digitar dois números e verifique se o primeiro é maior que o segundo. Continue pedindo números até que o primeiro número seja maior que o segundo.

while True:
    try:
        primeiro_numero = float(input("Digite o primeiro número: "))
        segundo_numero = float(input("Digite o segundo número: "))
       
        if primeiro_numero > segundo_numero:
            print("O primeiro número é maior que o segundo. Programa encerrado.")
            break
        else:
            print("O primeiro número não é maior que o segundo.")
    
    except ValueError:
        print("Digite números válidos.")
```

```python
# 60 Crie um programa que peça ao usuário um número e exiba todos os divisores desse número.

while True:
    try:
        numero = int(input("Digite um número para ver seus divisores: "))
        break  
    except ValueError:
        print("Por favor, digite um número inteiro válido.")

print(f"Divisores de {numero}:")
for i in range(1, numero + 1):
    if numero % i == 0:
        print(i)
```

```python
# 63 Desenvolva um algoritmo que peça ao usuário para inserir 5 números, adicione-os a uma lista e, depois, exiba a soma de todos os números na lista.

numeros = []

for i in range(5):
    while True:
        try:
            numero = float(input(f"Digite o {i + 1}º número: "))
            numeros.append(numero)
            break
        except ValueError:
            print("Digite um número válido.")

soma = sum(numeros)
print(f"\nA soma de todos os números é: {soma}")
```

```python
# 64  Crie uma lista com 10 números aleatórios e exiba apenas os números que são múltiplos de 3.
import random

numeros = [random.randint(1, 100) for _ in range(10)]

print("Números aleatórios:", numeros)
print("Múltiplos de 3:")
for numero in numeros:
    if numero % 3 == 0:
        print(numero)
```

```python
# 65 Escreva um programa que solicite ao usuário uma lista de 5 números e exiba o maior e o menor número dessa lista.

numeros = []

for i in range(5):
    while True:
        try:
            numero = float(input(f"Digite o {i + 1}º número: "))
            numeros.append(numero)
            break
        except ValueError:
            print("Digite um número válido.")

maior_numero = max(numeros)
menor_numero = min(numeros)

print(f"\nO maior número da lista é: {maior_numero}")
print(f"O menor número da lista é: {menor_numero}")
```

```python
# 66. Escreva um algoritmo que forneça um menu para o usuário: 1-Cadastrar nome, 2- atualizar o nome, 3, excluir, 4-listar todos os cadastrados, 
# ao final da operação deve dar uma mensagem indicando o resultado da operação e perguntando se deseja realizar
# outra operação, o seu aplicativo apenas deve encerrar quando a opção não for inserida.

nomes_cadastrados = []

def cadastrar_nome():
    nome = input("Digite o nome a ser cadastrado: ")
    nomes_cadastrados.append(nome)
    print(f"Nome '{nome}' cadastrado com sucesso!")

def atualizar_nome():
    nome_antigo = input("Digite o nome que deseja atualizar: ")
    if nome_antigo in nomes_cadastrados:
        novo_nome = input("Digite o novo nome: ")
        nomes_cadastrados[nomes_cadastrados.index(nome_antigo)] = novo_nome
        print(f"Nome atualizado de '{nome_antigo}' para '{novo_nome}'.")
    else:
        print(f"O nome '{nome_antigo}' não foi encontrado!")

def excluir_nome():
    nome = input("Digite o nome a ser excluído: ")
    if nome in nomes_cadastrados:
        nomes_cadastrados.remove(nome)
        print(f"Nome '{nome}' excluído com sucesso!")
    else:
        print(f"O nome '{nome}' não foi encontrado!")

def listar_nomes():
    if nomes_cadastrados:
        print("\nLista de nomes cadastrados:")
        for nome in nomes_cadastrados:
            print(nome)
    else:
        print("Não há nomes cadastrados.")

def exibir_menu():
    while True:
        print("\nMenu:")
        print("1 - Cadastrar nome")
        print("2 - Atualizar nome")
        print("3 - Excluir nome")
        print("4 - Listar todos os cadastrados")
        print("0 - Sair")
        
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            cadastrar_nome()
        elif opcao == '2':
            atualizar_nome()
        elif opcao == '3':
            excluir_nome()
        elif opcao == '4':
            listar_nomes()
        elif opcao == '0':
            print("Aplicativo encerrado.")
            break
        else:
            print("Opção inválida!")

        continuar = input("\nDeseja realizar outra operação? (s/n): ").lower()
        if continuar != 's':
            print("Aplicativo encerrado.")
            break

# Inicia o menu
exibir_menu()



```
