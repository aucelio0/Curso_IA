começando o tratamento de dados

Dataset1

```python
df_c11 = df_c1.copy()
```

```python
df_c11.info()
```

```python
df_c11.drop('nome', axis=1, inplace=True)
```

```python
df_c11.drop('data_inscricao', axis=1, inplace=True)
```

```python
df_c11.drop('contratado', axis=1, inplace=True)
```

```python
df_c11
```

```python
df_c11 = df_c11.rename(columns={
    'experiencia_anos': 'tempo_experiencia',
    'disponibilidade_viagem': 'aceita_viagem',
    'salario_vaga' : 'salario',
    'disponibilidade_mudanca': 'aceita_mudanca',
    'genero' : 'sexo'
})
```

```python
df_c11
```

Dataset2

```python
df_c22 = df_c2.copy()
```

```python
df_c22.info()
```

```python
df_c22.drop('data_cadastro', axis=1, inplace=True)
```

```python
df_c22.drop('nome_completo', axis=1, inplace=True)
```

```python
df_c22.drop('foi_contratado', axis=1, inplace=True)
```

```python
df_c22 = df_c22.rename(columns={
    'codigo_candidato': 'id_candidato',
    'uf' : 'estado',
    'faixa_etaria ' : 'idade',
    'nivel_educacional ': 'formacao',
    'setor_profissional' : 'area_atuacao',
    'qtd_certificacoes' : 'certificacoes',
    'idioma_principal' : 'idioma',
    'salario_esperado' : 'pretensao_salarial',
    'salario_oferecido' : 'salario',
    'deficiencia' : 'pcd',
    'idioma' : 'idiomas',
    'nivel_educacional' : 'formacao',
    'faixa_etaria' : 'idade'
})
```
